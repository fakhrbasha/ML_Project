```python
# Bar Chart
df.head(2)
```

```python
top_revenue = df.groupby('original_title').sum().sort_values(by="revenue" ,ascending = False)
top_revenue.head()
```

```python
movies = top_revenue.head().index
revenue = top_revenue.head().revenue.values
revenue
```

```python
plt.bar(movies , revenue);
```

```python
#pie chart
top_genres = df.groupby('genres').sum().sort_values(by =  'budget' , ascending = False)['budget'][0:5]
top_genres
```

```python
plt.pie(top_genres.values , labels = top_genres.index);
```
